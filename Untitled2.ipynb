```python
from transformers import BertTokenizer,BertForPreTraining
import torch
```

```python
with open("/content/final_doc_bert_all.txt","r") as t:
  pre_text=t.read().split("\n")
```

```python
pre_text
```

```python
text=[sentence for para in pre_text for sentence in para.split('.') if sentence!='']
```

```python
text
```

```python
c_size=len(text)
```

```python
c_size
```

```python
import random
sentence_a=[]
sentence_b=[]
labels=[]

for paragraph in text:
  sentences=[sentence for sentence in paragraph.split() if sentence!='']
  num_sentences=len(sentences)
  if num_sentences>1:
    sel= random.randint(0,num_sentences-2)
    sentence_a.append(sentences[sel])

    if random.random()>.5:
      sentence_b.append(text[random.randint(0,c_size-1)])
      labels.append(1)
    else:
      sentence_b.append(sentences[sel+1])
      labels.append(0)

      
```

```python
sentences
```

```python
tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
inputs=tokenizer(sentence_a,sentence_b,return_tensors='pt',max_length=128,truncation=True,padding='max_length')
```

```python
inputs['labels']=inputs.input_ids.detach().clone()
inputs['next_sentence_labels']=torch.LongTensor([labels]).T
```

```python
rand=torch.rand(inputs.input_ids.shape)
```

```python
mask=(rand<.15)*(inputs.input_ids!=101)*(inputs.input_ids!=102)
```

```python
sel=[]
for i in range(mask.shape[0]):
  sel.append(torch.flatten(mask[0].nonzero()).tolist())
```

```python
for i in range(mask.shape[0]):
  inputs.input_ids[i,sel[i]]=103
```

```python
class Srimad():
    def __init__(self, encodings):
        self.encodings = encodings
    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
    def __len__(self):
        return self.encodings.input_ids.shape[0]
```

```python
dataset=Srimad(inputs)
```

```python
loader = torch.utils.data.DataLoader(dataset, batch_size=16, shuffle=True)
```

```python
device=torch.device('cuda') 
```

```python
device
```

```python
model=BertForPreTraining.from_pretrained('bert-base-uncased')
```

```python
model.to(device)
```

```python
model.train()
```

```python
from transformers import AdamW
optimizer=AdamW(model.parameters(),lr=5e-2)
```

```python
from tqdm import tqdm  

epochs = 10

for epoch in range(epochs):
    loop = tqdm(loader, leave=True)
    for batch in loop:
        
        optimizer.zero_grad()
      
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids=batch['token_type_ids'].to(device)
        next_sentence_labels=batch['next_sentence_labels'].to(device)
        labels=batch['labels'].to(device)
        outputs = model(input_ids=input_ids,token_type_ids=token_type_ids, attention_mask=attention_mask,next_sentence_label=next_sentence_labels,labels=labels)
    
        loss = outputs.loss
        
        loss.backward()
      
        optimizer.step()

        loop.set_description(f'Epoch {epoch}')
        loop.set_postfix(loss=loss.item())
```
